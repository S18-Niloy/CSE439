```python
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

# Define the root folder where the dataset is located

```python
root_folder = r'C:\Users\Bijoy\Desktop\NILOY Desktop\CSE439\Data'
```

# Function to load images and masks

```python
def load_dataset(root_folder):
    images = []
    masks = []
    for tile_folder in os.listdir(root_folder):# Define the root folder where the dataset is located
        if tile_folder.startswith('Tile'):
            image_folder = os.path.join(root_folder, tile_folder, 'images')
            mask_folder = os.path.join(root_folder, tile_folder, 'masks')
            for filename in os.listdir(image_folder):
                if filename.endswith('.jpg'):
                    # Load image
                    image_path = os.path.join(image_folder, filename)
                    image = cv2.imread(image_path)
                    images.append(image)
                    # Load mask
                    mask_filename = filename.replace('.jpg', '.png')
                    mask_path = os.path.join(mask_folder, mask_filename)
                    mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)
                    masks.append(mask)
    return images, masks
```

# Load images and masks

```python
images, masks = load_dataset(root_folder)
```

```python
print(len(images))
print(len(masks))
```

# Visualize sample image and mask

```python
sample_index = 71  # Change this index to visualize different samples
sample_image = images[sample_index]
sample_mask = masks[sample_index]
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(sample_image, cv2.COLOR_BGR2RGB))
plt.title('Sample Image')
plt.subplot(1, 2, 2)
plt.imshow(sample_mask, cmap='gray')
plt.title('Sample Mask')
plt.show()
```
